```python
from pathlib import Path
import pandas as pd

data_path = Path("../data/raw/student_droput.csv")
if not data_path.exists():
    data_path = Path("../data/raw/student_dropout.csv")


df = pd.read_csv(data_path, sep=";")

print("Dataset shape:", df.shape)
df.head()
```

```python
print("Dataset shape:", df.shape)

print("\nColumn names:")
print(df.columns.tolist())
```

```python
print("Missing values:")
print(df.isnull().sum())

print("\nTotal missing values:", df.isnull().sum().sum())
print("Duplicate rows:", df.duplicated().sum())
```

```python
print("Target distribution:")
print(df["Target"].value_counts())

print("\nTarget percentage:")
print(df["Target"].value_counts(normalize=True) * 100)
```

```python
print("Data types:")
print(df.dtypes.value_counts())

df.dtypes
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.replace("\t", "", regex=False)
    .str.replace(" ", "_")
    .str.replace("/", "_")
    .str.replace("(", "", regex=False)
    .str.replace(")", "", regex=False)
    .str.lower()
)

print(df.columns.tolist())
```

```python
df_cleaned = df.copy()

df_cleaned.columns = (
    df_cleaned.columns
    .str.strip()
    .str.replace("\t", "", regex=False)
    .str.replace(" ", "_")
    .str.replace("/", "_")
    .str.replace("(", "", regex=False)
    .str.replace(")", "", regex=False)
    .str.replace("'", "", regex=False)
    .str.lower()
)

df_binary = df_cleaned[df_cleaned["target"].isin(["Dropout", "Graduate"])].copy()

df_binary["dropout_risk"] = df_binary["target"].map({
    "Dropout": 1,
    "Graduate": 0
})

df_binary = df_binary.drop(columns=["target"])

print("Binary dataset shape:", df_binary.shape)
print(df_binary["dropout_risk"].value_counts())
print(df_binary["dropout_risk"].value_counts(normalize=True) * 100)
```

```python
from pathlib import Path

project_root = Path.cwd()

processed_path = project_root / "data" / "processed"
processed_path.mkdir(parents=True, exist_ok=True)

df_binary.to_csv(processed_path / "student_dropout_cleaned.csv", index=False)

print("Saved to:", processed_path / "student_dropout_cleaned.csv")
```
